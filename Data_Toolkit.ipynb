```python
#  How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr = np.array([[1, 2, 3, 4], [7, 8, 9, 10], [11, 12, 13, 14]])
arr.ndim
print("2D  NumPy array", arr)
row_sum = np.sum(arr, axis=1)
print("The row sum is", row_sum)
```

```python
 # How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Math": [88, 92, 80, 89, 100, 67, 78, 85, 90, 94],
    "Science": [84, 90, 70, 85, 95, 65, 75, 80, 88, 92],
    "English": [78, 85, 75, 82, 89, 60, 72, 80, 86, 90],
    "History": [82, 88, 72, 84, 91, 62, 74, 81, 87, 93]
}

df = pd.DataFrame(data)
corr_matrix = df.corr()
print("Correlation Matrix:\n", corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#  Generate a bar plot using Plotly
import plotly.express as px
import pandas as pd
data = {
    "Fruit": ["Apple", "Banana", "Orange", "Grapes", "Mango"],
    "Quantity": [10, 15, 7, 20, 12]
}

df = pd.DataFrame(data)
fig = px.bar(
    df,
    x="Fruit",
    y="Quantity",
    color="Fruit",
    title="Fruit Quantity Bar Plot"
)

fig.show()
```

```python
# Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 1]
y3 = [5, 7, 3, 8, 2]

plt.plot(x, y1, color = "red", marker = "o", label="line y1")
plt.plot(x, y2, color = "blue", marker = "o", label="line y2")
plt.plot(x, y3, color = "green", marker = "o", label="line y3")
plt.title("Multiple lines in one plot")

plt.xlabel("Days")
plt.ylabel("stock price")
plt.legend()

plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'], 'Age': [25, 30, 35, 40], 'Salary': [50000, 60000, 75000, 80000], 'Incentives': [15000, 20000, 12000, 16000]}
df = pd.DataFrame(data)
print(df)
threshold = 50000
filtered_df = df[df["Salary"] > threshold]
print("\nFiltered DataFrame (Salary > 50000):\n", filtered_df)
```

```python
# Perform matrix multiplication using NumPy.
import numpy as np
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])
print("Matrix A:", A)
print("Matrix B:", B)
C = A @ B  # or C = A @ B
print("Result of A x B:", C)
```
